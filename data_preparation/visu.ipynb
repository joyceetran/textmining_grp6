```python
filename = "../datasets/raw_mda/_extraction_summary.csv"
documentcountdata = pd.read_csv(filename)
```

```python
documentcountdata.head()
```

```python
company_summary = (
    documentcountdata
    .groupby("company_name", as_index=False)
    .agg(
        unique_ciks=("cik", "nunique"),
        filings_found=("filings_found", "sum"),
        filings_saved=("filings_saved", "sum"),
        mda_extracted=("mda_extracted", "sum")
    )
    .sort_values("company_name")
)

print(company_summary)
print(f"Total unique companies: {documentcountdata['company_name'].nunique()}")
```

```python
total_filings_found = documentcountdata["filings_found"].sum()
total_filings_saved = documentcountdata["filings_saved"].sum()

print(f"Total filings found: {total_filings_found}")
print(f"Total filings saved: {total_filings_saved}")
```
