```python
# Load files with pickle

import pickle

with open("nb_model.pkl", "rb") as f:
    model = pickle.load(f)

with open("nb_vectorizer.pkl", "rb") as f:
    vectorizer = pickle.load(f)

with open("positive_words.pkl", "rb") as f:
    positive_words = pickle.load(f)

with open("negative_words.pkl", "rb") as f:
    negative_words = pickle.load(f)
```

```python
# Use models as inference

import re
import pandas as pd
from scipy.sparse import hstack, csr_matrix

def clean_text(text):
    text = str(text).lower().strip()
    text = re.sub(r"\s+", " ", text)
    return text

def lm_count_features(text):
    words = re.findall(r"\b\w+\b", str(text).lower())
    pos_count = sum(word in positive_words for word in words)
    neg_count = sum(word in negative_words for word in words)
    return pd.DataFrame([{
        "lm_pos_count": pos_count,
        "lm_neg_count": neg_count
    }])

def predict_sentiment(text):
    cleaned = clean_text(text)
    text_vec = vectorizer.transform([cleaned])
    lm_feats = lm_count_features(cleaned)
    X_final = hstack([text_vec, csr_matrix(lm_feats.values)])
    return model.predict(X_final)[0]
```

```python
# Example usage
sample_text = "the company reported strong demand and improved revenue this quarter"

result = predict_sentiment(sample_text)
print(result)
```
